ANALISE DE DADOS - KAGGLE

```python
# Importar as bibliotecas

import numpy as np
import pandas as pd
```

```python
# Carregar o arquivo 

base = pd.read_csv('agricultural_raw_material.csv')
```

```python
# Mostra a tabela

display(base)
```

```python
# Mostrar as informações da base

base.info()
```

```python
# Verificar se tem valores nulos ou ausentes

base.isnull().sum()
```

```python
# fazendo as correções com dados ausentes e incorretos ( usando Replace  "%"  ","  "-")

base = base.replace('%', '', regex=True)
base = base.replace(',', '', regex=True)
base = base.replace('-', '', regex=True)
base = base.replace('', np.nan)
base = base.replace('MAY90', np.nan)
```

```python
# Excluir os valores nulos e ausentes

base = base.dropna()
```

```python
# Verificar se ja apagou os valores nulos ou ausentes

base.isnull().sum()
```

```python
# Mostrando a informações após exluir os valores nulos ou ausentes.

base.info()
```

```python
# Mostrando tipos de dados

base.dtypes
```

```python
# Podemos converter os tipos das colunas para FLOAT

list = ['Coarse wool Price',
        'Coarse wool price % Change',
        'Copra Price',
        'Copra price % Change',
        'Cotton price % Change',
        'Fine wool Price',
        'Fine wool price % Change',
        'Hard log price % Change',
        'Hard sawnwood price % Change',
        'Hide price % change',
        'Plywood price % Change',
        'Rubber price % Change',
        'Softlog price % Change',
        'Soft sawnwood price % Change',
        'Wood pulp price % Change'
        ]

base[list] = base[list].astype('float')
```

```python
# Mostra  se está atualizados todos os tipos para FLOAT

base.dtypes
```

```python
base.head()
```

```python
# Corrigir a data e horas nas colunas

base.Month = pd.to_datetime(base.Month.str.upper(), format= '%b%y', yearfirst=False)

base = base.set_index('Month')
```

```python
# Visualizar a tabela com data corrigida na coluna ( Month )

base.head()
```

Análise Exploratória e Visualização de dados.

    - Matplotlib
    - Seaborn

```python
# Importando as bibliotecas

import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Configurando as informações para melhor visual

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize'] = (9, 5)
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

```python
# Colocando a lista de colunas que vão ser usada para mapa de calor ( correlação ) entre os preços.

list_mat = ['Coarse wool Price',  'Copra Price','Cotton Price', 'Fine wool Price',  'Hard log Price',
            'Hard sawnwood Price', 'Hide Price', 'Plywood Price', 'Rubber Price', 'Softlog Price', 
            'Soft sawnwood Price', 'Wood pulp Price'
            ]

corr_mat = base[list_mat].corr()
```

```python
# Mostrar gráfico

fig = plt.figure(figsize = (12, 9))

sns.heatmap(corr_mat, vmax = .8, square = True, annot = True)
plt.show()
```

```python
# Outro grafico diferente para fazer a comparação entre os preços

fig = plt.figure(figsize = (12, 9))

mask = np.triu(np.ones_like(list_mat, dtype=bool))

sns.heatmap(corr_mat, vmax = .8,mask=mask, square = True, annot = True)
plt.show()
```

```python
# Mostrando a correlação entre as matérias-primas em maior valor correlacionado e terem chance de ser duas matérias-primas. (Change)

change_list=['Copra price % Change','Softlog price % Change','Rubber price % Change','Cotton price % Change',
            'Coarse wool price % Change','Fine wool price % Change','Hard log price % Change','Hard sawnwood price % Change',
            'Hide price % change','Plywood price % Change','Soft sawnwood price % Change','Wood pulp price % Change'
            ]

# Novas correlações da lista nova
new_corr = base[change_list].corr()

# Mostrando os gráficos
plt.figure(figsize=(30,15))

new_corr= base[change_list].corr()
sns.heatmap(new_corr, annot=True)
plt.show()
```

```python
plt.figure(figsize=(7, 7))
lowlist=['Cotton Price','Hide Price','Softlog Price','Hard log Price','Soft sawnwood Price','Rubber Price']
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.subplots_adjust( hspace=0.5 )
    sns.boxplot(x=base[lowlist[i]])
```

```python
plt.figure(figsize=(7, 7))
highlist=['Coarse wool Price','Copra Price','Fine wool Price','Hard sawnwood Price','Plywood Price','Wood pulp Price']
for i in range(6):
    plt.subplot(2, 3, i+1)
    plt.subplots_adjust( hspace=0.5 )
    sns.boxplot(x=base[highlist[i]])
```

```python
plt.figure(figsize=(10, 10))

plt.plot(base[['Copra Price','Softlog Price','Rubber Price','Cotton Price','Coarse wool Price','Fine wool Price','Hard log Price',
             'Hard sawnwood Price','Hide Price','Plywood Price','Soft sawnwood Price','Wood pulp Price']]
         )

plt.title("Comparação de preços de matérias-primas")

plt.xlabel('Years')
plt.ylabel('Prices')

plt.legend(['Copra Price','Softlog Price','Rubber Price','Cotton Price','Coarse wool Price','Fine wool Price','Hard log Price',
            'Hard sawnwood Price','Hide Price','Plywood Price','Soft sawnwood Price','Wood pulp Price'],
           loc='upper center',bbox_to_anchor=(1.2,1))
```
